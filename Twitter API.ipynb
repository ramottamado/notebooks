```python
import oauth2
import json
import numpy as np
import pandas as pd

import IPython.display as display

CONSUMER_KEY = "PLACE CONSUMER KEY HERE"
CONSUMER_SECRET = "PLACE CONSUMER SECRET HERE"
ACCESS_TOKEN = "PAKE ACCESS TOKEN SENDIRI DUNKZ"
ACCESS_SECRET = "GA PUNYA? BIKIN"

# aing ngambil 200 tweet terbaru, ganti parameter count kalo mau lebih
URL = 'https://api.twitter.com/1.1/statuses/home_timeline.json?count=200'

def oauth_req(url, key=CONSUMER_KEY, secret=CONSUMER_SECRET, http_method="GET",
              post_body=b"", http_headers=None):
    consumer = oauth2.Consumer(key, secret)
    token = oauth2.Token(key=ACCESS_TOKEN, secret=ACCESS_SECRET)
    client = oauth2.Client(consumer, token)
    resp, content = client.request(url, method=http_method,
                                   body=post_body, headers=http_headers)
    return content

data = oauth_req(URL).decode('utf8')
jsondata = json.loads(data)
userdata = [tweets['user']['screen_name'] for tweets in jsondata]
tweetdata = [tweets['text'] for tweets in jsondata]

#datas = np.array([userdata, tweetdata])
#datas
data_pd = pd.DataFrame({'User': userdata, 'Tweet': tweetdata})
data_pd
```
